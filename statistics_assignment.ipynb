Q1:  What is a random variable in probability theory?

Ans: In probability theory, a *random variable* is a variable that represents the outcome of a random experiment. It's a way to assign numerical values to outcomes in a probability space.

There are two main types:

1. *Discrete random variable* – Takes on a countable number of possible values (e.g., number of heads in 3 coin flips).
2. *Continuous random variable* – Takes on an infinite number of possible values within a range (e.g., the time it takes for a bus to arrive).

*Example:*

- Rolling a six-sided die:  
  Let \( X \) be a random variable representing the outcome.  
  \( X \in \{1, 2, 3, 4, 5, 6\} \)

Even though the word "variable" is used, a random variable isn't random once the outcome is known—it just maps outcomes to numbers in a structured way.

Q4: What are probability distribution functions (PDF)?


Ans: In probability theory, a *Probability Distribution Function (PDF)* describes how the probability of a random variable is distributed. However, the term "PDF" *specifically refers to continuous random variables*.

Here’s a breakdown:

---

### *1. For Continuous Random Variables*
The *Probability Density Function (PDF)* is a function, usually denoted \( f(x) \), that:

- Describes the *relative likelihood* that a random variable takes on a particular value.
- The *area under the curve* of the PDF over an interval gives the *probability* that the variable falls in that interval.

#### Key Properties:
- \( f(x) \geq 0 \) for all \( x \)
- The total area under the curve is 1:  
  \[
  \int_{-\infty}^{\infty} f(x)\, dx = 1
  \]
- \( P(a \leq X \leq b) = \int_a^b f(x)\, dx \)

> Note: For continuous variables, *\( P(X = x) = 0 \)*. Only intervals matter.

---

### *2. For Discrete Random Variables*
While not a PDF, the analogous concept is the *Probability Mass Function (PMF)*, which assigns probabilities directly to specific values:

- \( P(X = x) = p(x) \), where \( \sum p(x) = 1 \)


Q5: How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?


Ans: Here's a clear comparison between *PDF* and *CDF*, and how they differ:

---

### *PDF (Probability Density Function)*
- *Applies to:* Continuous random variables
- *Purpose:* Describes how *densely* the probability is packed around each value.
- *Gives:* The *relative likelihood* of a value occurring.
- *Mathematical use:*  
  You *integrate the PDF* over an interval to get the probability:  
  \[
  P(a \leq X \leq b) = \int_a^b f(x)\, dx
  \]
- *Example:* Bell curve (normal distribution)

---

### *CDF (Cumulative Distribution Function)*
- *Applies to:* Both discrete and continuous random variables
- *Purpose:* Describes the *total probability* accumulated up to a certain value.
- *Gives:* The *probability that \( X \) is less than or equal to a value \( x \)*:  
  \[
  F(x) = P(X \leq x)
  \]
- *Relationship to PDF:*  
  The CDF is the *integral* of the PDF:  
  \[
  F(x) = \int_{-\infty}^x f(t)\, dt
  \]
  Conversely, the PDF is the *derivative* of the CDF:  
  \[
  f(x) = \frac{d}{dx}F(x)
  \]

---

### Quick Visual Analogy:
- *PDF: Think of it like the **speed* of filling a container.
- *CDF: Think of it like the **total amount filled* up to a point.



Q6: What is a discrete uniform distribution?


Ans: A *discrete uniform distribution* is a type of *discrete probability distribution* where *all possible outcomes are equally likely*.

---

### *Definition:*
If a random variable \( X \) can take on *\( n \)* distinct values:  
\[
X \in \{x_1, x_2, x_3, ..., x_n\}
\]  
and each value has the *same probability*, then:
\[
P(X = x_i) = \frac{1}{n} \quad \text{for all } i
\]

---

### *Key Properties:*
- All outcomes have *equal probability*
- Used when there is *no bias* toward any outcome
- *Mean (Expected Value):*  
  \[
  E[X] = \frac{a + b}{2}
  \]
- *Variance:*  
  \[
  \text{Var}(X) = \frac{(b - a + 1)^2 - 1}{12}
  \]  
  where \( a \) and \( b \) are the smallest and largest values

---

### *Example:*
Rolling a fair 6-sided die:  
\[
X \in \{1, 2, 3, 4, 5, 6\}
\]
Each outcome has probability:  
\[
P(X = x) = \frac{1}{6}
\]

---



Q7: What are the key properties of a Bernoulli distribution?


Ans: The *Bernoulli distribution* is one of the simplest and most important distributions in probability and statistics. It models a single trial that has *only two possible outcomes*: success or failure.

---

### *Key Properties of a Bernoulli Distribution:*

| Property | Description |
|----------|-------------|
| *Type* | Discrete |
| *Possible Values* | \( X \in \{0, 1\} \) (0 = failure, 1 = success) |
| *Parameter* | \( p \): the probability of success (where \( 0 \leq p \leq 1 \)) |
| *PMF (Probability Mass Function)* |  
\[
P(X = x) =
\begin{cases}
p & \text{if } x = 1 \\
1 - p & \text{if } x = 0
\end{cases}
\] |
| *Mean (Expected Value)* | \( E[X] = p \) |
| *Variance* | \( \text{Var}(X) = p(1 - p) \) |
| *Skewness* | \( \frac{1 - 2p}{\sqrt{p(1 - p)}} \) |

---

### *Example:*
- Tossing a coin:  
  Let "head" = 1 (success), "tail" = 0 (failure)  
  If the coin is fair, \( p = 0.5 \)

Q8: What is the binomial distribution, and how is it used in probability?


Ans: The *binomial distribution* is a discrete probability distribution that models the number of *successes* in a fixed number of *independent Bernoulli trials, each with the **same probability of success*.

---

### *Definition:*
Let \( X \) be the number of successes in \( n \) trials, where each trial has:
- *Success probability* \( p \)
- *Failure probability* \( 1 - p \)

Then \( X \sim \text{Binomial}(n, p) \), and the *probability mass function (PMF)* is:
\[
P(X = k) = \binom{n}{k} p^k (1 - p)^{n - k}
\]
Where:
- \( \binom{n}{k} \) is "n choose k" = number of ways to choose \( k \) successes from \( n \) trials
- \( k \in \{0, 1, 2, ..., n\} \)

---

### *Key Properties:*

| Property | Formula |
|---------|---------|
| *Mean (Expected Value)* | \( E[X] = np \) |
| *Variance* | \( \text{Var}(X) = np(1 - p) \) |
| *Type* | Discrete |

---

### *Use in Probability:*

The binomial distribution is used when you:
- Repeat a *fixed number* of independent trials
- Each trial has only *two outcomes* (success/failure)
- The *probability of success stays the same* for each trial

---



Q9: What is the Poisson distribution and where is it applied?


Ans: The *Poisson distribution* is a discrete probability distribution that models the *number of times an event occurs* in a fixed interval of *time or space, **given a constant average rate* and *independent occurrences*.

---

### *Definition:*

A random variable \( X \sim \text{Poisson}(\lambda) \), where:
- \( \lambda \) is the *average number of occurrences* in the interval (rate)
- \( X \in \{0, 1, 2, 3, \dots\} \)

*Probability Mass Function (PMF):*
\[
P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!}
\]
for \( k = 0, 1, 2, \dots \)

---

### *Key Properties:*

| Property | Formula |
|----------|---------|
| *Mean (Expected Value)* | \( E[X] = \lambda \) |
| *Variance* | \( \text{Var}(X) = \lambda \) |
| *Type* | Discrete |

---

### *When to Use the Poisson Distribution:*
Use it when you're counting *how many times* something happens in a fixed period of *time, distance, area, or volume*, and:
- Events occur *independently*
- The average rate \( \lambda \) is *constant*
- Two events *can’t happen at the exact same instant* (events are rare or spaced)

---



Q10: What is a continuous uniform distribution?


Ans: The *continuous uniform distribution* is a type of probability distribution where a *continuous random variable* is *equally likely to take any value* within a specific interval \([a, b]\).

---

### *Definition:*
A random variable \( X \) has a *continuous uniform distribution* on the interval \([a, b]\), written as:  
\[
X \sim \text{Uniform}(a, b)
\]

The *Probability Density Function (PDF)* is:
\[
f(x) =
\begin{cases}
\frac{1}{b - a} & \text{for } a \leq x \leq b \\
0 & \text{otherwise}
\end{cases}
\]

---

### *Key Properties:*

| Property | Formula |
|---------|---------|
| *Mean (Expected Value)* | \( E[X] = \frac{a + b}{2} \) |
| *Variance* | \( \text{Var}(X) = \frac{(b - a)^2}{12} \) |
| *CDF (Cumulative Distribution Function)* |  
\[
F(x) =
\begin{cases}
0 & x < a \\
\frac{x - a}{b - a} & a \leq x \leq b \\
1 & x > b
\end{cases}
\] |

---

### *Visual Idea:*
The PDF is a *flat horizontal line, meaning the probability is **evenly spread* across the interval.

Q11:  What are the characteristics of a normal distribution?

Ans: The *normal distribution, also known as the **Gaussian distribution, is one of the most important and widely used probability distributions in statistics. It describes data that clusters around a **mean* (average) and is symmetrically distributed.

---

### *Key Characteristics of a Normal Distribution:*

| Feature | Description |
|--------|-------------|
| *Shape* | Bell-shaped and symmetric around the mean |
| *Mean = Median = Mode* | All located at the center of the distribution |
| *Symmetry* | The left and right halves are mirror images |
| *Spread* | Determined by the *standard deviation (\( \sigma \))* |
| *Total Area Under Curve* | Equal to *1* (represents 100% probability) |
| *Tails* | Extend infinitely in both directions but never touch the axis |
| *68–95–99.7 Rule (Empirical Rule)* | About *68%* of data falls within 1σ, *95%* within 2σ, and *99.7%* within 3σ of the mean |

---

### *Probability Density Function (PDF):*
\[
f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{ -\frac{1}{2} \left( \frac{x - \mu}{\sigma} \right)^2 }
\]
Where:
- \( \mu \) = mean (center of the distribution)
- \( \sigma \) = standard deviation (controls the spread)

---

### *Applications:*
- Heights, weights, test scores
- Measurement errors
- Natural phenomena (e.g., IQ scores, blood pressure)

---

### *Standard Normal Distribution:*
A special case where \( \mu = 0 \) and \( \sigma = 1 \). It's used to calculate *z-scores*.

---



Q12: What is the standard normal distribution, and why is it important?


Ans: The *standard normal distribution* is a *special case* of the normal distribution where:

- *Mean* \( \mu = 0 \)
- *Standard deviation* \( \sigma = 1 \)

It is denoted as:
\[
Z \sim \mathcal{N}(0, 1)
\]

---

### *Why is it Important?*

1. *Simplifies Calculations:*  
   Probabilities for any normal distribution can be transformed into the standard normal form using *z-scores*, which makes it easier to use tables or software.

2. *Z-score Conversion:*  
   Any normal random variable \( X \sim \mathcal{N}(\mu, \sigma^2) \) can be standardized as:
   \[
   Z = \frac{X - \mu}{\sigma}
   \]
   This converts \( X \) into the standard normal distribution.

3. *Used in Hypothesis Testing and Confidence Intervals:*  
   Many statistical tests (like z-tests) and confidence interval formulas are built on the assumption of standard normality.

4. *Widely Tabulated:*  
   Standard normal distribution tables (z-tables) provide cumulative probabilities \( P(Z \leq z) \), which are essential for finding probabilities and critical values.

Q13: What is the Central Limit Theorem (CLT), and why is it critical in statistics?


Ans: The *Central Limit Theorem (CLT)* is one of the most powerful and fundamental theorems in statistics.

---

### *Definition of the Central Limit Theorem:*

The CLT states that:

> *When you take the mean of a large number of independent, identically distributed (i.i.d.) random variables, the sampling distribution of the sample mean will tend to follow a normal distribution — regardless of the original population's distribution.*

---

### *In Simple Terms:*
If you:
- Take repeated random samples from *any population*
- Each sample is of size \( n \)
- The population has a finite mean \( \mu \) and standard deviation \( \sigma \)

Then, as \( n \) becomes *large (usually \( n \geq 30 \)), the distribution of the **sample mean \( \bar{X} \)* will approach a *normal distribution* with:

\[
\text{Mean of } \bar{X} = \mu,\quad \text{Standard Deviation of } \bar{X} = \frac{\sigma}{\sqrt{n}}
\]

---

### *Why Is the CLT Important?*

1. *Justifies Using the Normal Distribution*  
   Even if the original data isn’t normally distributed, we can use *normal-based methods* (like z-tests) for inference on the mean.

2. *Enables Hypothesis Testing & Confidence Intervals*  
   Since sample means tend to be normally distributed, we can build *confidence intervals* and perform *statistical tests* using normal probabilities.

3. *Foundation for Inferential Statistics*  
   It allows statisticians to make predictions and generalizations from samples to populations — a cornerstone of statistics.

---

Q15: What is the application of Z statistics in hypothesis testing?


Ans: *Z-statistics* (or *z-tests) are used in **hypothesis testing* when you're working with *normally distributed data* or when the *sample size is large* and the *population standard deviation is known*.

---

### *Purpose of Z-Statistic:*
The z-statistic helps you determine *how far* a sample statistic (like the sample mean) is from the population parameter under the *null hypothesis, in terms of **standard deviations*.

---

### *Formula for Z-statistic (for sample mean):*
\[
z = \frac{\bar{x} - \mu}{\sigma / \sqrt{n}}
\]
Where:
- \( \bar{x} \) = sample mean  
- \( \mu \) = population mean (under the null hypothesis)  
- \( \sigma \) = population standard deviation  
- \( n \) = sample size

---

### *Steps in Hypothesis Testing using Z-statistic:*

1. *State the hypotheses:*
   - Null hypothesis \( H_0: \mu = \mu_0 \)
   - Alternative hypothesis \( H_1: \mu \neq \mu_0 \), \( \mu < \mu_0 \), or \( \mu > \mu_0 \)

2. *Choose significance level \( \alpha \)* (e.g., 0.05)

3. *Compute the Z-statistic*

4. *Find the critical value(s)* from the *standard normal table* (z-table)

5. *Make a decision:*
   - If \( |z| \) is greater than the critical value, *reject \( H_0 \)*
   - Otherwise, *fail to reject \( H_0 \)*

---

### *When to Use a Z-Test:*
- Large sample size (typically \( n \geq 30 \))
- Population standard deviation \( \sigma \) is known
- Testing means or proportions

---



Q16: How do you calculate a Z-score, and what does it represent?


Ans: A *Z-score* tells you *how many standard deviations* a data point is from the *mean* of a distribution.

---

### *Z-Score Formula:*

For a *single value* from a population:
\[
z = \frac{x - \mu}{\sigma}
\]

For a *sample mean*:
\[
z = \frac{\bar{x} - \mu}{\sigma / \sqrt{n}}
\]

Where:
- \( x \) = individual data point  
- \( \bar{x} \) = sample mean  
- \( \mu \) = population mean  
- \( \sigma \) = population standard deviation  
- \( n \) = sample size  

---

### *What the Z-score Represents:*

| Z-score | Interpretation |
|--------|----------------|
| 0 | Exactly at the mean |
| +1 | 1 standard deviation *above* the mean |
| -1 | 1 standard deviation *below* the mean |
| +2 or -2 | 2 standard deviations from the mean, etc. |

---

### *Why It’s Useful:*
- *Standardizes* different distributions for comparison
- Helps *calculate probabilities* using the standard normal distribution
- Used in *hypothesis testing, **grading curves, and **outlier detection*

---



Q17: What are point estimates and interval estimates in statistics?


Ans: In statistics, *point estimates* and *interval estimates* are both used to make inferences about a population based on sample data — but they serve different purposes.

---

### *1. Point Estimate:*

A *point estimate* is a *single value* used to estimate a population parameter.

- *Example:*  
  If you take a sample and calculate the mean \( \bar{x} = 75 \), this is your *point estimate* of the population mean \( \mu \).

#### *Common Point Estimates:*
| Parameter | Point Estimate |
|-----------|----------------|
| Population Mean \( \mu \) | Sample Mean \( \bar{x} \) |
| Population Proportion \( p \) | Sample Proportion \( \hat{p} \) |
| Population Variance \( \sigma^2 \) | Sample Variance \( s^2 \) |

---

### *2. Interval Estimate:*

An *interval estimate* gives a *range of values* (called a confidence interval) within which the population parameter is likely to fall, along with a *confidence level*.

- *Example:*  
  A 95% confidence interval for the mean might be \( [72, 78] \), meaning you're 95% confident the true mean is between 72 and 78.

#### *Confidence Interval for a Mean (known \( \sigma \)):*
\[
\bar{x} \pm z^* \cdot \frac{\sigma}{\sqrt{n}}
\]

Where:
- \( \bar{x} \) = sample mean  
- \( z^* \) = z-value corresponding to the confidence level (e.g., 1.96 for 95%)  
- \( \sigma \) = population standard deviation  
- \( n \) = sample size

---

### *Key Differences:*

| Feature | Point Estimate | Interval Estimate |
|--------|----------------|-------------------|
| *Form* | Single number | Range of numbers |
| *Uncertainty* | Doesn’t express it | Shows uncertainty |
| *Interpretation* | Best guess | Range of likely values |

---



Q18: What is the significance of confidence intervals in statistical analysis?  


Ans: *Confidence intervals (CIs)* are crucial in statistical analysis because they provide *more information* than a single point estimate. They show both the *estimate* and the *degree of uncertainty*, helping us make more reliable conclusions about a population.

---

### *Key Significance of Confidence Intervals:*

#### 1. *Range of Likely Values*
- A CI gives a *range* where the *true population parameter* (like a mean or proportion) is likely to fall.
- For example: “We are 95% confident that the average test score is between 68 and 72.”

---

#### 2. *Incorporates Uncertainty*
- All samples have *sampling error. A confidence interval reflects this by giving a **margin of error* around the point estimate.

---

#### 3. *Helps in Decision Making*
- CIs allow you to make informed decisions — especially when comparing groups or testing hypotheses.
- Example: If a 95% CI for the difference between two means doesn’t include 0, the difference is statistically significant.

---

#### 4. *More Informative than P-values Alone*
- While a p-value tells you if an effect is significant, a CI shows you *how big the effect is* and *how precise* the estimate is.

---

#### 5. *Used Across Many Fields*
- Whether it's polling, medicine, quality control, or economics, CIs are widely used to *generalize sample results* to populations.

---



Q19: What is the relationship between a Z-score and a confidence interval?


Ans: Great question! The *Z-score* and *confidence interval (CI)* are closely related — the Z-score helps *determine the margin of error* when constructing a CI, especially when using the *normal distribution*.

---

### *The Relationship:*

When you're creating a *confidence interval for a population mean* (and the population standard deviation \( \sigma \) is known), the formula is:

\[
\text{Confidence Interval} = \bar{x} \pm z^* \cdot \frac{\sigma}{\sqrt{n}}
\]

Where:
- \( \bar{x} \) = sample mean  
- \( \sigma \) = population standard deviation  
- \( n \) = sample size  
- \( z^* \) = *Z-score* corresponding to the desired confidence level

---

### *Z-Score Values for Common Confidence Levels:*

| Confidence Level | Z-Score (\( z^* \)) |
|------------------|----------------------|
| 90%              | 1.645                |
| 95%              | 1.96                 |
| 99%              | 2.576                |

---

### *What Does the Z-Score Do in a CI?*
- The Z-score determines *how wide* the confidence interval is.
- Higher confidence → larger \( z^* \) → *wider interval* (more certainty, less precision)
- Lower confidence → smaller \( z^* \) → *narrower interval* (more precision, less certainty)

---



Q20: How are Z-scores used to compare different distributions?


Ans: *Z-scores* are super useful when you want to *compare values* from *different distributions, especially when those distributions have **different means and standard deviations*.

---

### *How Z-scores Help in Comparison:*

The Z-score standardizes a value by converting it into the *number of standard deviations* it is away from the mean:

\[
z = \frac{x - \mu}{\sigma}
\]

Where:
- \( x \) = value from the dataset  
- \( \mu \) = mean of the dataset  
- \( \sigma \) = standard deviation of the dataset

This allows you to *compare across different units or scales, because Z-scores are unitless and all follow the **standard normal distribution*.

---

### *Example:*

Suppose two students take different tests:

- *Student A* scores *85* on a math test where  
  \( \mu = 80 \), \( \sigma = 5 \)

- *Student B* scores *78* on a physics test where  
  \( \mu = 70 \), \( \sigma = 4 \)

Let’s find Z-scores:

- \( z_A = \frac{85 - 80}{5} = 1.0 \)
- \( z_B = \frac{78 - 70}{4} = 2.0 \)

*Conclusion:*  
Even though Student A scored higher numerically, *Student B performed better relative to their peers, because their score is **2 standard deviations above the mean*.

Q21: What are the assumptions for applying the Central Limit Theorem?



Ans: The *Central Limit Theorem (CLT)* is powerful, but it relies on some *key assumptions* to hold true. Here are the main ones:

---

### *1. Independent Random Samples*
- The observations must be *independent* of each other.
- This means the selection of one value *does not affect* the selection of another.

---

### *2. Identically Distributed Variables*
- The random variables should be *identically distributed, meaning they all come from the **same population* with the same probability distribution.

---

### *3. Finite Mean and Standard Deviation*
- The population must have a *finite mean* \( \mu \) and *finite standard deviation* \( \sigma \).  
- CLT doesn’t apply well if the variance is infinite (like with Cauchy distributions).

---

### *4. Large Sample Size*
- The sample size \( n \) should be *reasonably large* (typically *\( n \geq 30 \)*).
- If the original population is *normally distributed*, CLT works even with small samples.
- For *skewed or non-normal distributions*, a larger \( n \) may be needed.

---

### *Summary Table:*

| Assumption                  | Why It Matters                          |
|----------------------------|-----------------------------------------|
| Independence               | Ensures no correlation between samples |
| Identical distribution     | Keeps the mean and variance consistent |
| Finite variance            | Prevents unstable or infinite spread   |
| Sufficiently large sample  | Allows the sampling distribution to be approx. normal |

---



Q22: What is the concept of expected value in a probability distribution?


Ans: The *expected value* (also called *mean* or *expectation) of a probability distribution represents the **average outcome* you would expect *over many repetitions* of a random experiment.

---

### *Definition:*

The expected value is a *weighted average* of all possible outcomes, where each outcome is weighted by its *probability*.

---

### *1. For Discrete Random Variables:*

\[
E(X) = \sum [x_i \cdot P(x_i)]
\]

Where:
- \( x_i \) = possible values of the random variable  
- \( P(x_i) \) = probability of each value

---

### *2. For Continuous Random Variables:*

\[
E(X) = \int_{-\infty}^{\infty} x \cdot f(x)\,dx
\]

Where:
- \( f(x) \) = probability density function (PDF)

---



### *Why It’s Important:*
- Helps in *decision making* and *risk analysis*

Q23: How does a probability distribution relate to the expected outcome of a random variable?


Ans:

A *probability distribution* tells you *how likely* each possible outcome of a random variable is — and from that, you can determine the *expected outcome, or **expected value*.

---

### *How They Relate:*

- The *probability distribution* gives the *set of outcomes* and their *associated probabilities*.
- The *expected value* is a *summary measure* of the distribution — it’s the *long-run average* of the outcomes, weighted by their probabilities.

---

### *Mathematically:*

For a *discrete random variable*:

\[
E(X) = \sum [x_i \cdot P(x_i)]
\]

Each value \( x_i \) of the random variable is multiplied by its *probability* \( P(x_i) \), and the sum gives the expected outcome.

```python
def bernoulli_pmf(x, p):
    """
    Calculate the PMF of a Bernoulli distribution.

    Parameters:
    x : int (0 or 1) - The outcome
    p : float - Probability of success (P(X=1))

    Returns:
    float - Probability of the outcome x
    """
    if x not in [0, 1]:
        return 0  # Invalid outcome for Bernoulli
    return p if x == 1 else 1 - p

# Example usage
p = 0.7  # probability of success
print("P(X = 0):", bernoulli_pmf(0, p))
print("P(X = 1):", bernoulli_pmf(1, p))
```

Q15: Implement hypothesis testing using Z-statistics for a sample dataset?

```python
import numpy as np
from scipy.stats import norm

# Step 1: Sample data
sample_data = [102, 98, 105, 100, 95, 99, 101, 97, 96, 103]
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)  # sample standard deviation
n = len(sample_data)

# Step 2: Population mean under null hypothesis
mu_0 = 100

# Step 3: Calculate Z-statistic
z = (sample_mean - mu_0) / (sample_std / np.sqrt(n))

# Step 4: Calculate two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z)))

# Step 5: Decide significance level
alpha = 0.05

# Step 6: Print results
print(f"Sample Mean = {sample_mean:.2f}")
print(f"Sample Std Dev = {sample_std:.2f}")
print(f"Z-Statistic = {z:.2f}")
print(f"P-value = {p_value:.4f}")

if p_value < alpha:
    print("Reject the null hypothesis (H₀).")
else:
    print("Fail to reject the null hypothesis (H₀).")
```

Q16: Create a confidence interval for a dataset using Python and interpret the result?

```python
import numpy as np
from scipy.stats import norm

def confidence_interval(data, confidence=0.95):
    """
    Compute the confidence interval for a given dataset.

    Parameters:
    data : list or array-like - Sample dataset
    confidence : float - Confidence level (default is 95%)

    Returns:
    (lower_bound, upper_bound) - Confidence interval
    """
    # Sample statistics
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # Sample standard deviation
    n = len(data)

    # Z-score for the given confidence level
    alpha = 1 - confidence
    z_score = norm.ppf(1 - alpha/2)  # Two-tailed

    # Margin of error
    margin_of_error = z_score * (sample_std / np.sqrt(n))

    # Confidence Interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound, sample_mean

# Example dataset
sample_data = [102, 98, 105, 100, 95, 99, 101, 97, 96, 103]

# Compute the confidence interval
lower, upper, mean = confidence_interval(sample_data)

# Display results
print(f"Sample Mean = {mean:.2f}")
print(f"95% Confidence Interval = ({lower:.2f}, {upper:.2f})")

# Interpretation
print("\nInterpretation:")
print(f"We are 95% confident that the true population mean lies between {lower:.2f} and {upper:.2f}.")
```

Q24: Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal  distribution.?

```python
import numpy as np
import scipy.stats as stats

def compute_skewness_kurtosis(size=1000, mu=50, sigma=10):
    """
    Generates a dataset from a normal distribution and computes skewness & kurtosis.

    Parameters:
    - size: Number of samples
    - mu: Mean of the normal distribution
    - sigma: Standard deviation of the normal distribution
    """
    # Step 1: Generate normally distributed data
    np.random.seed(42)
    data = np.random.normal(mu, sigma, size)

    # Step 2: Calculate skewness and kurtosis
    skewness = stats.skew(data)
    kurtosis = stats.kurtosis(data, fisher=True)  # Fisher=True for excess kurtosis

    # Print results
    print(f"Skewness: {skewness:.4f}")
    print(f"Kurtosis: {kurtosis:.4f}")

    return skewness, kurtosis

# Call the function
compute_skewness_kurtosis()
```
